```python
import seaborn as sns
import matplotlib.pyplot as plt
import PyPDF2 as PyPDF2
import nltk
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize, sent_tokenize
import string
from rake_nltk import Rake
import pandas as pd


#================Summary===========================
mypdf = open("corpus/19660006049.pdf", mode='rb')
pdfReader = PyPDF2.PdfFileReader(mypdf)
text=''
for i in range(0,pdfReader.numPages):
    # creating a page object
    pageObj = pdfReader.getPage(i)
    # extracting text from page
    text=text+pageObj.extractText()
#tokenize
stopWords = set (stopwords.words("english"))
words = word_tokenize(text)
#frequency table
freqTable = dict()
for word in words:
    word = word.lower()
    if word in stopWords:
        continue
    if word in freqTable:
        freqTable[word] +=1
    else:
        freqTable[word] =1
#store sentences
sentences = sent_tokenize(text)
sentenceValue = dict()
for sentence in sentences:
    for word, freq in freqTable.items():
        if word in sentence.lower():
            if sentence in sentenceValue:
                sentenceValue[sentence] +=1
            else:
                sentenceValue[sentence] =1

sumValues = 0
for sentence in sentenceValue:
    sumValues += sentenceValue[sentence]
average = int(sumValues / len(sentenceValue))
#store sentences in summary
summary = ''
for sentence in sentences:
    if (sentence in sentenceValue) and (sentenceValue[sentence] > (3 * average)):
        summary += " " + sentence
print()
print("====This is the Summary===")
print(summary)
#================Keywords===========================
rake_nltk_var = Rake()
#remove punctuation
text = text.translate(str.maketrans('', '', string.punctuation))
#rake
rake_nltk_var.extract_keywords_from_text(text)
keyword_extracted = rake_nltk_var.get_ranked_phrases()
# Creating FreqDist with 20 most common tokens
all_fdist = nltk.FreqDist(keyword_extracted).most_common(20)

# Convert to Pandas for plotting
all_fdist = pd.Series(dict(all_fdist))

## Setting figure, ax into variables
fig, ax = plt.subplots(figsize=(10,30))

print()
print("===Keyword list===")
print(all_fdist)

#================Keywords Frequency plot===========================
print()
print("===Keyword Frequency plot===")
## Seaborn plotting using Pandas attributes + xtick rotation for ease of viewing
all_plot = sns.barplot(y=all_fdist.index, x=all_fdist.values, ax=ax)
plt.xticks(rotation=30);




```
